```python
def is_parallelogram (x1, y1, x2, y2, x3, y3, x4, y4):
  mid1_x, mid1_y =(x1+x3)/2, (y1+y3)/2
  mid2_x, mid2_y =(x2+x4)/2, (y2+y3)/2
  if(mid1_x==mid2_x) and (mid1_y==mid2_y):
    return True
  if((x2-x1,y2-y1)==(x4-x3,y4-y3) and (x3-x2,y3-y2)==(x1-x4,y1-y4)):
    return True
  return False
x1,y1=map(int,input('a1=').split())
x2,y2=map(int,input('a2=').split())
x3,y3=map(int,input('a3=').split())
x4,y4=map(int,input('a4=').split())
if is_parallelogram(x1, y1, x2, y2, x3, y3, x4, y4):
  print("  ")
else:
  print("   ")
```
