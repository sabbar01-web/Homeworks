```python
def C(n, m):
    if m == 0 or m == n:
        return 1
    return C(n - 1, m) + C(n - 1, m - 1)
n = int(input("Введите n: "))
m = int(input("Введите m: "))
if 0 <= m <= n:
    print(f"C({n}, {m}) =", C(n, m))
else:
    print("Ошибка: должно быть 0 <= m <= n")
```
