```python
def sum_series (epsilon=0.001):
  S=0
  n=1
  while True:
    a_n=(2*n-1)/(2**n)
    if abs(a_n)<epsilon:
      break
    S+=a_n
    n+=1
  return S
print(" =",sum_series())
```
