While3.

```python
N = int(input("n="))
K = int(input("k="))
quotient = 0
remainder = N
while remainder >= K:
    remainder -= K
    quotient += 1
print(f"Частное: {quotient}, Остаток: {remainder}")
```

While8.

```python
N = int(input("n="))
K = 0
while (K + 1) ** 2 <= N:
    K += 1
print(f"Наибольшее K, при котором K^2 <= N: {K}")
```

While20.

```python
N = int(input("n="))
has_two = False
while N > 0:
    digit = N % 10
    if digit == 2:
        has_two = True
        break
    N = N // 10
print(has_two)
```
